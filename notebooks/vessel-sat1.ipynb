```python
import rasterio
import numpy as np

DIR = '/home/dakk/sattest/'
DS = 'S2B_MSIL2A_20230428T100559_N0509_R022_T32SNJ_20230428T133214.SAFE/GRANULE/L2A_T32SNJ_A032080_20230428T101606/IMG_DATA/'
# Open the band files using rasterio
b02 = rasterio.open(DIR + DS + 'R10m/T32SNJ_20230428T100559_B02_10m.jp2')
b03 = rasterio.open(DIR + DS + 'R10m/T32SNJ_20230428T100559_B03_10m.jp2')
b04 = rasterio.open(DIR + DS + 'R10m/T32SNJ_20230428T100559_B04_10m.jp2')
scl = rasterio.open(DIR + DS + 'R20m/T32SNJ_20230428T100559_SCL_20m.jp2')

# Read the data into numpy arrays
b02_data = b02.read(1).astype(np.uint16)
b03_data = b03.read(1).astype(np.uint16)
b04_data = b04.read(1).astype(np.uint16)
```

```python
from rasterio.enums import Resampling

cloud_mask = scl.read(1, out_shape=(b02_data.shape[0], b02_data.shape[1]), resampling=Resampling.nearest)
cloud_mask_bool = np.isin(cloud_mask, [3,8,9,10], invert=False)

# b02_data = np.ma.masked_where(cloud_mask_bool, b02_data)
# b03_data = np.ma.masked_where(cloud_mask_bool, b03_data)
# b04_data = np.ma.masked_where(cloud_mask_bool, b04_data)
```

```python
# Create a new rasterio dataset with the desired output format
with rasterio.open('/home/dakk/output.tif', 'w', driver='GTiff', 
                   width=b02.width, height=b02.height,
                   count=3, dtype=b02_data.dtype, 
                   crs=b02.crs, transform=b02.transform) as dst:
    # Write the data to the new dataset
    dst.write(b02_data, 3)
    dst.write(b03_data, 2)
    dst.write(b04_data, 1)
```
